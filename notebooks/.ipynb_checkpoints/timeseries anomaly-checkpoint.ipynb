```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import torch.nn as nn
```

```python
df_small_noise_url_suffix = "../data/artificialNoAnomaly/artificialNoAnomaly/art_daily_small_noise.csv"
df_small_noise_url = df_small_noise_url_suffix
df_small_noise = pd.read_csv(
    df_small_noise_url, parse_dates=True, index_col="timestamp"
)

df_daily_jumpsup_url_suffix = "../data/artificialWithAnomaly/artificialWithAnomaly/art_daily_jumpsup.csv"
df_daily_jumpsup_url = df_daily_jumpsup_url_suffix
df_daily_jumpsup = pd.read_csv(
    df_daily_jumpsup_url, parse_dates=True, index_col="timestamp"
)
```

```python
# Normalize and save the mean and std we get,
# for normalizing test data.
training_mean = df_small_noise.mean()
training_std = df_small_noise.std()
df_training_value = (df_small_noise - training_mean) / training_std
print("Number of training samples:", len(df_training_value))
```

```python
TIME_STEPS = 288

# Generated training sequences for use in the model.
def create_sequences(values, time_steps=TIME_STEPS):
    output = []
    for i in range(len(values) - time_steps):
        output.append(values[i : (i + time_steps)])
    return np.stack(output)


x_train = create_sequences(df_training_value.values)
print("Training input shape: ", x_train.shape)
```

autoencoder class

```python
class ConvAutoEncoder(nn.Module):
    def __init__(self):
        #Encoder
        self.conv1 = nn.Conv1d(7, 32, 7, stride=2, padding=1)
        self.conv2 = nn.Conv1d(7, 16, 7, stride=2, padding=1)
        
        self.t_conv1 = nn.ConvTranspose1d(7, 16, 7, stride=2, padding=1)
        self.t_conv2 = nn.ConvTranspose1d(7, 32, 7, stride=2, padding=1)
```
